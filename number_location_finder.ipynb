```python
import numpy as np
import pandas as pd
import phonenumbers
from opencage.geocoder import OpenCageGeocode
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from phonenumbers import geocoder,carrier
import folium
```

```python
numbers = '+2347052575511'
```

```python
pepnumber = phonenumbers.parse(numbers)
location = geocoder.description_for_number(pepnumber, 'en')
print(location)
```

```python
#service_provider = phonenumbers.parse(numbers)
print(carrier.name_for_number(pepnumber, 'en'))
```

```python
key = '86665e8ce650448aa5990b336c2e2875'
geocoder = OpenCageGeocode(key)
result = geocoder.geocode(str(location))
print(result)
```

```python
lat = result[0]['geometry']['lat']
lng = result[0]['geometry']['lng']
print(lat, lng)
```

```python
myMap = folium.Map(location=[lat,lng], zoom_start=9)
folium.Marker([lat,lng], popup=location).add_to(myMap)
myMap.save("mymap.html")
```
